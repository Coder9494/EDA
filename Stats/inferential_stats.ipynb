### Z test using Python


```python
'''
Suppose the IQ in a certain population is normally distributed with mean of u = 100 and standard deviation of sigma = 15.
A researcher wants to know if a new drug affects IQ levels, so he recruits 20 patients to try it and record their IQ levels.

The following code shows how to perform a one sample z-test in Python to determine if the new drug causes a significant difference in IQ levels:
'''
from statsmodels.stats.weightstats import ztest as ztest

# enter IQ levels for 20 patients
data = [88, 92, 94, 96, 97, 99, 99, 98, 105, 109, 110, 112, 113, 113, 119, 120, 120, 120, 121, 124, 130 ]

ztest(data, value = 100)
```

```python
'''
1st parameter is the data
2nd parameter is the mean of the population = 100

we get 2 output
1 --> z-test value = 3.2353507202332894
2 --> p-value      = 0.0012149339639596398    ----> p value can be used along with significance value to accept or reject Null hypothesis
                                                    0.05 > 0.0012149  i.e. after comparing them, we can accept the Null Hypothesis
'''
```

#### t-test using Python

```python
## creating data

ages = [10,20,23,45,63,55,67,84,12,34,11,56,78,90,98,33,24,56,25,89,65,22,34,56,12,22,3,67,89,55,21,27,85,80,20,43,24,24,56,75,35,74,24, 23, 24, 23, 45, 44, 55, 65, 23 ,12]

import numpy as np
ages_mean = np.mean(ages)
ages_mean
```

```python
from scipy.stats import ttest_1samp
ttest_1samp(ages_sample, 43)

# check the significanve value with respect to pvalue and conclude on hypothesis testing 
```

```python
'''
Consider ages of the entire college students as population.
Ages mean of a class of student is given by below code
'''
import numpy as np
import pandas as pd
import scipy.stats as stats
import math

np.random.seed(6)
# the below code gives ages starting from 18, mean as 35 and size as 1500
school_ages = stats.poisson.rvs(loc = 18, mu = 35, size = 1500)
classA_ages = stats.poisson.rvs(loc = 18, mu = 30, size = 60)
```

```python
school_ages
```

```python
classA_ages
```

```python
classA_ages.mean()
```

```python
# applying t-test 
_, p_value = ttest_1samp(classA_ages, school_ages.mean())

if p_value <= 0.05:
    print("The null hypothesis is rejected using t-test", p_value)
else:
    print("the null hypothesis is accepted using t-test", p_value)

# the pvalue is 10 raise to -13 lower than significance value of alpha = 0.05, hence it needs to be rejected
```

#### Difference between p-value and significanve value

```python
'''
The average weight of all residents in Bangalore city is 168 pounds with a standard deviation 3.9.
We take a sample of 36 individuals and the mean is 169.5 pounds.
Confidence Interval is 95%age.

i.e. population mean = 168
sigma = 3.9
(sample mean) mean bar = 169.5

alpha = 0.05

check the decision boundary (-1.96 to 1.96)

get the z value using z-test formula
z = 2.307

Since z value is greater than 1.96, we reject the null hypothesis

but where does p-value comes into play?
with respect to z-score what is the area under the curve using z-table ---> 0.99111
i.e. based on z-score, the area under the curve is 0.99111
area outside the curve is 1- 0.99111 = .00889
area to left/ right = (0.00889)/2 = 0.0044

to get the p-value = 0.0044 + 0.0044 = 0.0088

from p-value is less than 0.05, so we reject the null hypothesis

Suppose if p-value is greater than significance value, we accept the null hypothesis


'''
```

### Covariance

```python
'''
It is used to quantify relationship between two variables
Ex1: Height    Weight     ---> here height increaseing, then weight also increasing
    150       56
    155       58
    159       60
    165       65
    170       72

Ex2: Play         Study   ---> here play time increasing, study time decreasing
     2            6
     3            5
     4            4
     5            3
     6            2
'''
```

Covariance is given by https://www.probabilitycourse.com/chapter5/5_3_1_covariance_correlation.php





This will give
1. +ve when X is increasing, Y is also increasing i.e. direct relationship
2. -ve when X is increasing, Y is decreasing i.e. indirect relationship
3. 0 when no relation between X and Y

Disadvantages of Covariance --> there is no fixed value of magnitude, i.e. we can see the direction but how much relationship is not given.

So, to restrict the values and know how much related variables are, we use Pearson Corelation Coeeficient


#### Pearson Corelation
- limits the values from (-1,1)
- the more the value is -1 or 1, the more related the two variables are.
- good prediction for linear relationship i.e. it captures the linear relationship very efficiently
- Checkout the wikipedia page --> https://en.wikipedia.org/wiki/Pearson_correlation_coefficient



#### Spearman's rank correlation coefficient
- checkout the wikipedia page - https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient
