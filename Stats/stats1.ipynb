#### Outliers

```python
# define our dataset
dataset = [11,12,4,2,9,7,4,7,5,6,7,0,9,11,4,109,3,2,6,7,9,3,100,12,9,120,10,8,2,5,3,7,4,11,1,6,0,3,2,1]

plt.hist(dataset, bins= 30)
```

```python
# function to find outliers - z-score computation
outliers = []

def detect_outliers(data):
    threshold = 3    # since 99.7 % percentage of data lies inside 3rd standard deviation (empirical formula)
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z_score = (i-mean)/std

        if np.abs(z_score) > threshold:    # np.abs is used to get absolute value i.e. rounding off
            outliers.append(i)

    return outliers
```

```python
# get the outliers

detect_outliers(dataset)
```

```python
# Inter-quartile range (IQR)
'''
1. Sort the data
2. Calculate Q1 and Q3
3. Substract Q3 - Q1 to find IQR
4. Find lower fence (Q1-1.5(IQR))
5. Find upper fence (Q3 + 1.5(IQR))
'''

# sorting the data
dataset = sorted(dataset)

# calculating Q1 and Q3
q1, q3 = np.percentile(dataset, [25,75])

# calculating IQR
iqr = q3 - q1

# compute lower fence
lower_fence = q1 - 1.5*iqr

# compute higher fence
higher_fence = q3 + 1.5*iqr

print(lower_fence, higher_fence)
```

```python
# write function to remove all the outliers using IQR


def remove_outliers(data):
    rmv_outliers = []
    for i in data:
        if i > lower_fence and i < higher_fence:
            rmv_outliers.append(i)
    return rmv_outliers

# calling the function on earlier created dataset
remove_outliers(dataset)
```

```python
# instead of doing all the above things for outlier, we can directly use below library

import seaborn as sns
sns.boxplot(dataset)

# the three outliers are visible
```
